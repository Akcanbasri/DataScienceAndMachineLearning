# Boxplot

## Veri setinin hikayesi  

**total_bill:** yemeğin toplam fiyatı (veri ve bahşiş dahil)  
<br>
**tip:** bahşiş  
<br>
**sex:** ücreti ödeyen kişinin cinsiyeti (kadın: 1, erkek: 0)  
<br>
**smoker:** grupta sigara içen bar mı (no: 0 , yes: 1)  
<br>
**day:** gün (3=perşembe, 4=cuma, 5=cumartesi, 6=pazar)  
<br> 
**time:** ne zaman? (gündüz=0, gece=1)  
<br>
**size:** grupta kaç kişi var?  

```python
# importing data set from seaborn library
import seaborn as sns

tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
df["sex"].value_counts()
```

```python
df["smoker"].value_counts()
```

```python
df["day"].value_counts()
```

```python
df["time"].value_counts()
```

# Boxplot

```python
sns.boxplot(x=df["total_bill"])
```

# caprazlamalar

sorumuz şu: Hangi gün daha fazla kazanıyoruz? 

```python
sns.boxplot(x="day", y="total_bill", data=df);
```

```python
# sabah mı aksam mı daha fazla kazanıyoruz?
sns.boxplot(x="time", y="total_bill", data=df);
```

```python
# kişi sayısı ile kazanç arasındaki ilişki
sns.boxplot(x="size", y="total_bill", data=df);
```

```python
# üç boyutlu grafik
sns.boxplot(x="day", y="total_bill", hue="sex", data=df);
 
```

# Violin Grafiği

```python
sns.catplot(y="total_bill", kind="violin", data=df);
```

## çarprazlama

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", data=df);
```

#### Heatmap

```python
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe().T
```

```python
pivot_df = df.pivot(index="month", columns="year", values="passengers")
```

```python
sns.heatmap(pivot_df, annot=False);
```
